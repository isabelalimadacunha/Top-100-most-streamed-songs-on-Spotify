# Importing necessary libraries, loading the dataset and renaming columns for clarity


```python
import pandas as pd

streamed_data = pd.read_csv("/content/drive/MyDrive/Spotify Data/Top 100 most Streamed - Sheet1.csv")
streamed_data.rename(columns = {'top genre':'genre'}, inplace = True)

# Displaying the initial dataset
display(streamed_data)
```

# Handling missing values and displaying a summary of the columns



```python
streamed_data = streamed_data.dropna()
display(streamed_data.info())

# Displaying descriptive statistics, rounded to one decimal place
display(streamed_data.describe().round(1))
```

# Counting the number and the percentage of occurrences of each artist

```python
qtde_artist = streamed_data["artist"].value_counts()
display(qtde_artist)

qtde_artist_perc = streamed_data["artist"].value_counts(normalize=True)
display(qtde_artist_perc)
```

# Visualizing the distribution of each column, colored by artist

```python
import plotly.express as px


for coluna in streamed_data:
  grafico = px.histogram(streamed_data, x=coluna, color="artist")
  grafico.show()
```
